```python
#Importing necessary libraries
import plotly.express as px
import pandas_profiling as pp
import pandas as pd
#Reading the COVID_19 dataset from the CSV file
with open("/content/Video games.csv", 'r') as file:
    data_frame = pd.read_csv(file)
data_frame.head()
```

```python
#exploring dataset
data_frame.info
```

```python
from sklearn.preprocessing import LabelEncoder

#Encoding columns using LabelEncoder
label_encoder = LabelEncoder()
data_frame["Name"] = label_encoder.fit_transform(data_frame["Name"])
data_frame["Platform"] = label_encoder.fit_transform(data_frame["Platform"])
data_frame["Genre"] = label_encoder.fit_transform(data_frame["Genre"])
data_frame["Publisher"] = label_encoder.fit_transform(data_frame["Publisher"])
```

```python
#Checking for missing values in the DataFrame
data_frame.isnull().sum()
```

```python
# Filling missing values in 'Age' column with the mode
data_frame['Year'].fillna(data_frame['Year'].mode()[0], inplace=True)
```

```python
# Filling missing values in 'Age' column with the mode
data_frame['Publisher'].fillna(data_frame['Publisher'].mode()[0], inplace=True)
```

```python
# Create the area chart
fig = px.area(data_frame, x='Year', y='Global_Sales', title='Global Sales Trend')
# Show the chart
fig.show()
```

```python
#create the histogram
fig = px.histogram(data_frame, x='Global_Sales', nbins=20)
#show the plot
fig.show()
```

```python
import plotly.graph_objs as go
import plotly.offline as pyo
```

```python
data_frame = pd.DataFrame(data_frame)

# Create the scatter plot
fig = px.scatter(data_frame, x='NA_Sales', y='EU_Sales', title='North American Sales vs. European Sales')

# Show the chart
fig.show()
```

```python
# Create the scatter plot
trace = go.Scatter(x=data_frame['JP_Sales'], y=data_frame['Other_Sales'], mode='markers', marker=dict(color=data_frame['Year']))

# Define the layout of the plot
layout = go.Layout(title='Scatter Plot: JAPAN vs Other Sales', xaxis=dict(title='JP_sales'), yaxis=dict(title='Other_sales'))

# Create the figure and add the trace
fig = go.Figure(data=[trace], layout=layout)

# Show the chart
fig.show()
```

```python
# stack the data
df =data_frame.set_index("Year").stack().reset_index()
# create the stacked area chart
fig = px.area(data_frame, x='Year', y='Global_Sales', color='Platform')
#show the chart
fig.show()
```

```python
# Get the total sales for each publisher
GRP1 = data_frame.groupby('Publisher', as_index=False)['Global_Sales'].sum()

# Sort the DataFrame by Global_Sales in descending order to get the top 10 publishers
df_sorted = GRP1.sort_values(by='Global_Sales', ascending=False).head(10)

# Create the bar chart
fig = px.bar(df_sorted, x='Publisher', y='Global_Sales', title='Top 10 Publishers by Global Sales')

# Show the chart
fig.show()
```

```python
# Get the total sales for each category
GRP2 = data_frame.groupby('Genre', as_index=False)['Global_Sales'].sum()

# Create the pie chart
fig = px.pie(GRP2, values='Global_Sales', names='Genre', title='Genre Distribution by Global Sales')

# Show the chart
fig.show()
```
