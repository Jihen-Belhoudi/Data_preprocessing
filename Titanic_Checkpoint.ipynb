### Instruction 2:
As a next step apply what you have learned in Data preprocessing to preprocess your data: find missing informations and replace them by the appropriate values.

We can infer from showing the summary of our dataframe using df.info() that we have missing values in 3 columns : **Age, Cabin and Embarked**
* for the Age, we will fill missing values with the median.
* for Cabin we have so much missing values, we need to drop the column from our dataframe.
* for the Embarked column we will fill missing values with the mode.

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df = df.drop('Cabin', axis=1)
```

We will utilize **LabelEncoder** for the Label columns **Survived**. For the features **Sex & Embarked** we will utilize **OneHotEncoder**.

```python
# Importing Transformer
from sklearn.preprocessing import LabelEncoder
```

```python
# Instanciate a label encoder
le = LabelEncoder()
# Encoding the 'Survived' Label
df['Survived'] = le.fit_transform(df['Survived'])
```

```python
df.head()
```

```python
df['Survived'].unique()
```

```python
# Get one hot encoding of columns Sex and Embarked
one_hot_s = pd.get_dummies(df['Sex'])
one_hot_e = pd.get_dummies(df['Embarked'])

# Drop columns Sex and Embarked as it is now encoded
df = df.drop(columns=['Sex', 'Embarked'])

# Join the encoded df
df = df.join([one_hot_s, one_hot_e])
```
