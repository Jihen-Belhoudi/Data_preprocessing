```python
import pandas as pd

data = {
    'Name': ['John', 'Mary', 'Bob', 'Sarah', 'Tom', 'Lisa'],
    'Department': ['IT', 'Marketing', 'Sales', 'IT', 'Finance', 'Marketing'],
    'Age': [30, 40, 25, 35, 45, 28],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Salary': [50000, 60000, 45000, 55000, 70000, 55000],
    'Experience': [3, 7, 2, 5, 10, 4]
}

employee_df = pd.DataFrame(data)

employee_df
```

```python
# Select the first 3 rows using iloc method
employee_df.iloc[:3]
```

```python
# Select all rows where the Department is "Marketing" using loc method
employee_df.loc[employee_df['Department'] == 'Marketing']
```

```python
# Select the Age and Gender columns for the first 4 rows of the dataframe using iloc method

# 1st method
employee_df.iloc[:5, [2, 3]]

# 2nd method
# employee_df.iloc[:5, [False, False, True, True, False, False]]
```

```python
# Select the Salary and Experience columns for all rows where the Gender is "Male" using loc method
employee_df.loc[employee_df['Gender'] == 'Male', ['Salary', 'Experience']]
```
